```python
import matplotlib.pyplot as plt
```

```python
x = ['Depth1', 'Depth2', 'Depth3']  # x轴数据
y = [0.5972, 0.5507, 0.5348] 
y_p = [0.6089, 0.5573, 0.5451]  
y_c = [0.6873, 0.6987, 0.7] 
y_pc = [0.6884, 0.6992, 0.7] 
```

```python
plt.plot(x, y, label='LLaVA-Geometry without prompt',linewidth=2,  marker='o')
plt.plot(x, y_p, label='LLaVA-Geometry with prompt',linewidth=2,  marker='o')
plt.plot(x, y_c, label='LLaVA-Geometry with Chain-of-Thought',linewidth=2,  marker='o')
plt.plot(x, y_pc, label='LLaVA-Geometry with prompt and Chain-of-Thought',linewidth=2,  marker='o')
plt.title('Accuracy under different difficulty levels and different training strategies')
plt.xlabel('Difficulty level')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--')
# 添加图例
plt.legend()
plt.savefig('train_visual.pdf')
plt.savefig('train_visual.png')
# 显示图形
plt.show()
```
