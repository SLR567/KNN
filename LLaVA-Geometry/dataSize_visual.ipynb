```python
import matplotlib.pyplot as plt

# 数据
labels = ['D1', 'D2', 'D3']
zero_answer_base = [0.3838, 0.3815, 0.3100]
zero_answer_base_prompt = [0.3975, 0.3930, 0.3775]
few_answer_base = [0.4853, 0.4689, 0.4607]
few_answer_base_prompt = [0.5130, 0.4858, 0.4747]
full_answer_base = [0.5972, 0.5507, 0.5348]
full_answer_base_prompt = [0.6089, 0.5573, 0.5451]

# 设置柱状图的宽度
bar_width = 0.2

# 计算每个柱状图的位置
index = [0, 1, 2]
index_zero = [i - bar_width for i in index]
index_few = index
index_full = [i + bar_width for i in index]

# 创建图形和轴对象
#fig, ax = plt.subplots()

# 绘制柱状图
plt.bar(index_zero, zero_answer_base_prompt, width=bar_width, label='Zero-shot with Prompt', alpha=1.0)
plt.bar(index_zero, zero_answer_base, width=bar_width, label='Zero-shot', alpha=1.0)
plt.bar(index_few, few_answer_base_prompt, width=bar_width, label='Few-shot with Prompt', alpha=1.0)
plt.bar(index_few, few_answer_base, width=bar_width, label='Few-shot', alpha=1.0)
plt.bar(index_full, full_answer_base_prompt, width=bar_width, label='Full data with Prompt', alpha=1.0)
plt.bar(index_full, full_answer_base, width=bar_width, label='Full data', alpha=1.0)


# 设置图形标题和轴标签
plt.title('Accuracy under different training data size and different test difficulty')
#plt.xlabel('Data')
plt.ylabel('Accuracy')

# 设置x轴刻度和标签
plt.xticks(index,labels)
#plt.xticklabels(labels)

# 添加图例
#plt.legend()

# 添加图例，并放在图的下方
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)

# 调整图形布局，以便图例不会超出图的边界
plt.tight_layout()

plt.savefig('dataSize_visual.pdf')
plt.savefig('dataSize_visual.png')
# 显示图形
plt.show()

```
