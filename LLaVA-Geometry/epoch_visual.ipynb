```python
import matplotlib.pyplot as plt

```

```python
# 数据
x = [0, 0.2, 0.4, 0.6, 0.8, 1]  # x轴数据
y1 = [0.3584, 0.5384, 0.5356, 0.5526, 0.5512,0.5609]  # 第一条数据的y轴数据
y2 = [0.3893, 0.5455, 0.5418, 0.5545, 0.5550,0.5704]  # 第二条数据的y轴数据


```

```python
# 创建图形和轴对象
#fig, ax = plt.subplots()

```

```python
# 绘制折线图
plt.plot(x, y1, label='LLaVA-Geometry without prompt',linewidth=2,  marker='o')
plt.plot(x, y2, label='LLaVA-Geometry with prompt',linewidth=2,  marker='o')

# 设置图形标题和轴标签
plt.title('Average accuracy under different epoches')
plt.xlabel('Epoch')
plt.ylabel('Average Accuracy')
plt.grid(True, linestyle='--')
# 添加图例
plt.legend()
plt.savefig('epoch_visual.pdf')
plt.savefig('epoch_visual.png')
# 显示图形
plt.show()
```
